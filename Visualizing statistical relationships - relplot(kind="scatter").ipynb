#### Scatter plot:
> * The scatter plot is a mainstay of statistical visualization. It depicts the joint distribution of two variables using a cloud of points, where each point represents an observation in the dataset. 
> * This depiction allows the eye to infer a substantial amount of information about whether there is any meaningful relationship between them.

```python
tips = sns.load_dataset("tips") # load tips dataset from seaborn library
```

```python
tips.head()
```

```python
tips.describe(include='all') # descriptive statistics of tips dataframe
```

```python
sns.relplot(x="total_bill", y="tip", kind="scatter", data=tips);  # kind="scatter" is default. You can modify it to kind="line"
```

Scatterplot is plotted with two dimensions but using seaborn, we can add one more dimension to the plot by coloring the points according to the thrird variable.In seaborn, this is referred to as using a <b>“hue semantic”.

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);plt.show()
```

To differentiate the classes, we can use different marker style for each class.

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker", data=tips);plt.show()
```

In Seaborn, we can represents four variables by changing the <b>hue & style</b> of each point independently.

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="sex", data=tips);plt.show()
```

When hue semantic is categorical then default <a href="https://seaborn.pydata.org/tutorial/color_palettes.html#palette-tutorial"> qualitative palette </a> is applied.
And when hue semantic is numeric (specifically, if it can be cast to float), then the default coloring switches to a sequential palette.

```python
sns.relplot(x="total_bill", y="tip", hue="size", data=tips);plt.show()
```

You can customize color palette.

```python
sns.relplot(x="total_bill", y="tip", hue="size", palette=sns.cubehelix_palette(rot=-.4), data=tips);plt.show()
```

Let's see how sizes parameter works:

```python
sns.relplot(x="total_bill", y="tip", size="size", sizes=(20, 200), palette=sns.cubehelix_palette(rot=-.9), data=tips);plt.show()
```
